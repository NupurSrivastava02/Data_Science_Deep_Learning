```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
dataset = pd.read_csv('sat.csv')
```

```python
dataset
```

```python
plt.scatter(dataset['SAT'],dataset['GPA'])
plt.show()
```

```python
X = dataset.iloc[:,0].values
```

```python
X
```

```python
y = dataset.iloc[:,1].values
```

```python
y
```

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([
    Dense(5,activation='relu',input_dim=1,kernel_regularizer=keras.regularizers.l2(0.001)),
    Dense(5,activation='relu'),
    Dense(5,activation='relu'),
    Dense(1)
    
])
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.0001))
```

```python
model.fit(X,y,epochs=500)
```

```python
plt.scatter(dataset['SAT'],dataset['GPA'])
plt.plot(X,model.predict(X),color='red')
plt.show()
```
