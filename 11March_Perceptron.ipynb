```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
X = np.array([-1,0,1,2,3,4])
```

```python
X
```

```python
y = np.array([-3,-1,1,3,5,7])
```

```python
y
```

```python
plt.scatter(X,y)
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
import statsmodels.api as sm
```

```python
x1 = sm.add_constant(X)
```

```python
x1
```

```python
model = sm.OLS(y,x1).fit()
```

```python
model.summary()
```

```python
# From above we can see b0 = -1 and b1 =2
```

```python
# We will apply this in Deep Learning for Perceptron i.e. single neural network
```

```python
X
```

```python
y
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([Dense(units=1,input_shape=[1])])
```

```python
model.summary()
```

```python
# Two parameters here are weight(w) and bias(b)
```

```python
#Loss will be calculated using the 'Mean_Squared_Error' for Regression in Deep Learning
```

```python
model.compile(loss='mean_squared_error',optimizer='sgd')
```

```python
model.fit(X,y,epochs=500)
```

```python
model.get_weights()
```

```python
# From the above we see bias(b) = -1, weight(w) = 2
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red')
plt.xlabel('independent')
plt.ylabel('Dependent')
plt.show()
```
